## Building Simple Recommender System

This project will using Weighted Average and based on genres. Use a formula from IMDB to calculate *Weighted_Average Rating (WR)*

$$ WR = (\frac{v}{v+m}.R ) + (\frac{m}{v+m}.C) $$
    
$v$: number of votes for the film <br>
$m$: minimum number of votes needed to be included in the chart <br>
$R$: average rating of the film <br>
$C$: average number of votes from the entire movie data <br>

```python
def imdb_weighted_rating(df, var=0.8):
    v = df['votes']
    R = df['users_rating']
    C = df['users_rating'].mean()
    m = df['votes'].quantile(var)
    df['score'] = (v/(m+v))*R + (m/(m+v))*C #IMDb formula 
    return df['score']
    
imdb_weighted_rating(movie_df)

#melakukan pengecekan dataframe
movie_df.head()
```

```python
df = movie_df.copy()

def user_prefer_recommender(df, ask_genre, top, var=0.8):
    m = df['votes'].quantile(var)
    
    if ask_genre.lower() == 'all':
        df = df
    else:
        def filter_genre(x):
            if ask_genre.lower() in str(x).lower():
                return True
            else:
                return False
        df = df.loc[df['genre'].apply(lambda x: filter_genre(x))]

    df = df.loc[df['votes'] >= m]
    df = df.sort_values(by='score', ascending=False)
    
    #get the top data
    df = df[:top]
    return df

user_prefer_recommender(df,
                        ask_genre = 'action',
                        top=5)
```
